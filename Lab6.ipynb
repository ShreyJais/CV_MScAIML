### Loading the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('/content/drive/MyDrive/Classroom/Data engineering and knowledge representation MSc AIML/1aad848b-3adc-4bcd-96f4-a0492d93d5c9.png',0)
plt.imshow(img,"gray")
```

### Median Filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= temp[4]
plt.imshow(img_new1,"gray")
```

### Max filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= temp[-1]
plt.imshow(img_new1,"gray")
```

### Min filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= temp[0]
plt.imshow(img_new1,"gray")
```

### Mid-point Filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= round((temp[0]+temp[-1])/2)
plt.imshow(img_new1,"gray")
```

### Alpha-trimmed Filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= round((sum(temp[1:-1]))/7)
plt.imshow(img_new1,"gray")
```
