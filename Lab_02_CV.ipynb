```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/Classroom/Data engineering and knowledge representation MSc AIML/1aad848b-3adc-4bcd-96f4-a0492d93d5c9.png')
```

```python
cv2_imshow(img)
```

```python
rows, cols, ch = img.shape
```

```python
M_translation = np.float32([[1, 0, 100],   # x translation
                            [0, 1, 50]])  # y translation
image_translation = cv2.warpAffine(img, M_translation, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_translation, cv2.COLOR_BGR2RGB)), plt.title('Translation')
plt.tight_layout()
plt.show()
```

```python
angle = 30
M_rotation = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
image_rotation = cv2.warpAffine(img, M_rotation, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_rotation, cv2.COLOR_BGR2RGB)), plt.title('Rotation')
plt.tight_layout()
plt.show()
```

```python
M_scaling = np.float32([[0.5, 0, 0],
                        [0, 0.5, 0]])
image_scaling = cv2.warpAffine(img, M_scaling, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_scaling, cv2.COLOR_BGR2RGB)), plt.title('Scaling')
plt.tight_layout()
plt.show()
```

```python
M_shearing = np.float32([[1, 0.5, 0],
                         [0.5, 1, 0]])
image_shearing = cv2.warpAffine(img, M_shearing, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_shearing, cv2.COLOR_BGR2RGB)), plt.title('Shearing')
plt.tight_layout()
plt.show()
```

```python
M_reflection_horizontal = np.float32([[-1, 0, cols],
                                      [0, 1, 0]])
image_reflection_horizontal = cv2.warpAffine(img, M_reflection_horizontal, (cols, rows))

plt.figure(figsize=(12, 12))
plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_reflection_horizontal, cv2.COLOR_BGR2RGB)), plt.title('Reflection (Horizontal)')
plt.tight_layout()
plt.show()
```
